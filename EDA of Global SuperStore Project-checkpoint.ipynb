### Import Required Libraries:
   Import pandas, numpy, matplotlib, and seaborn
   Set up matplotlib to display plots inline


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### Load the Dataset

```python
df = pd.read_csv('Global_Superstore.csv')
```

```python
df.head(20)
```

#### Data Exploration

```python
# Check the shape of the DataFrame
print('Data Frame of Data set is',df.shape)
print('\n')
# Display info about the DataFrame using `info()`
print('Information about Data set is',df.info())
print('\n')
# Generate descriptive statistics using `describe()`
print('Descriptive Statistics is:\n',df.describe())
```

#### Data Cleaning


```python
# Check for missing values
print(df.isnull().sum())
# Handle any missing or inconsistent data
df.dropna(inplace=True)
# Convert data types if necessary (e.g., dates to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

```python
# After clean data looks
df.info()
df
df.to_csv("cleaned_data.csv")
```

#### Basic Analysis

```python
#   Calculate total sales and profit
Total_Sales = df['Sales'].sum()
Total_Profit = df['Profit'].sum()

print(f'Total Sales : {Total_Sales} & Total Profit : {Total_Profit}\n')

# Identify top-selling products

top_product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print('--------Top selling products---------\n',top_product)

# Analyze sales by region
print('\n')
sales_region = df.groupby('Region')['Sales'].sum()
print('--------Top selling region---------\n', sales_region)

# Examine the relationship between sales and profit
print('\n')
correlation = df['Sales'].corr(df['Profit'])
print(f'Correlation betweeen Sales and Profit: {correlation:.3f}')
```

#### Data Visulization

```python
# Create a bar chart of sales by product category
sns.barplot(x='Category',y='Sales',data=df)
plt.title('Sales by Product Category')
plt.show()
```

```python
# Plot a line graph of monthly sales over time
df['Month'] = df['Order Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Sales'].sum() # Monthly Sales

monthly_sales.index = monthly_sales.index.to_timestamp()

monthly_sales.plot(kind='line')
plt.title('Monthly Sales Over Time')
plt.show()
```

```python
# Generate a scatter plot of sales vs. profit
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs. Profit')
plt.show()
```

```python
# Produce a heatmap of sales by region and product category

pivot_table = df.pivot_table(values = 'Sales', index='Region', columns='Category', aggfunc='sum')

sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlGnBu' )
plt.title('Sales by Region and Product Category')
plt.show()
```

### Step 8 Derive Insights

```python
# Seasonal Decomposition of time Series (monthly sales)
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(monthly_sales, model='additive', period=12)
result.plot()
plt.show()
```

### Step 9 Bonus Challenge

   Perform a simple time series analysis to identify any sales trends or seasonality

```python
# convert 'Discount' and 'Profit' columns to numeric

df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# calculate correlation between discount and profit 
discount_profit_correlation = df[['Discount','Profit']].corr().iloc[0,1]

total_discount = df['Discount'].sum()
total_profit = df['Profit'].sum()

print("Total Discount:", total_discount)
print("Total Profit:",total_profit)
print("Correlation between Discount and Profit:\n", discount_profit_correlation)
```
